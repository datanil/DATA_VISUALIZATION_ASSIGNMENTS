### **Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

### ***ANSWER :***

#### ***Here are five commonly used plots in Seaborn along with their uses:***

* **Scatter Plot:**

 > Scatter plots are used to visualize the relationship between two numerical variables. Each point on the plot represents an individual data point, making it useful for identifying patterns, clusters, or correlations between the variables.

* **Bar Plot:**

> Bar plots are used to display and compare the distribution of categorical data. They are suitable for showing the frequency, counts, or proportions of different categories in the data.

* **Histogram:**

> Histograms are used to represent the distribution of a single numerical variable. It shows how the data is spread across different bins, helping to identify the underlying data distribution and any outliers.

* **Box Plot (Box-and-Whisker Plot):**

> Box plots are excellent for visualizing the distribution of numerical data through quartiles. They provide a compact way to show the median, quartiles, and possible outliers in the data.

* **Heatmap:**

> Heatmaps are used to visualize the correlation matrix between multiple numerical variables. They use color intensity to represent the strength of the relationship between variables, making it easier to identify patterns and dependencies.
These are just a few examples, and Seaborn offers many more types of plots that can be used for different visualization tasks. It's a versatile library that allows you to create appealing and informative visualizations with relatively simple code.








### **Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using**
### **x ="timepoint" and y = "signal" for different events and regions.**


### **Note: timepoint, signal, event, and region are columns in the fmri dataset.**

### ***ANSWER :***

```python
import seaborn as sns
fmri  =  sns.load_dataset('fmri')

sns.lineplot(x = fmri.timepoint , y = fmri.signal)
```

### **Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using**
### **x = 'pclass', y = 'age' and y = 'fare'.**


### **Note: pclass, age, and fare are columns in the titanic dataset.**

### ***ANSWER :***

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic =  sns.load_dataset('titanic')

fig , axes =  plt.subplots(2,1 , figsize=(8,10))

sns.boxplot(x = titanic.pclass , y = titanic.age , ax = axes[0])
axes[0].set_title('Pclass Vs Age')


sns.boxplot(x = titanic.pclass , y = titanic.fare , ax= axes[1])
axes[1].set_title("Pclass vs Fare")
```

### **Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

### ***ANSWER :***

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

sns.displot(data = diamonds , x = 'price' , hue = 'cut' )
plt.title('Histogram of Diamond Price cut')

```

### **Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**

### **Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

### ***ANSWER :***

```python
import seaborn as sns

flights = sns.load_dataset('flights' )

flights.head()
```

```python
flight = flights.pivot('month','year','passengers')

plt.figure(figsize=(10,8))
sns.heatmap(flight ,annot = True , fmt = 'd')
```
