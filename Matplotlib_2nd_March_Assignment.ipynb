### **Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

### ***ANSWER  :***

* **Matplotlib :**
> Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of tools and functions for creating various types of plots and charts. Matplotlib is widely used in scientific computing, data analysis, and data visualization tasks.

* **Five plots that can be plotted using the Pyplot module of Matplotlib:**

 *  **Line Plot:**
> A line plot is used to represent data points connected by straight lines. It is commonly used to visualize trends or patterns over time or any ordered variable.

 * **Scatter Plot:**
> A scatter plot displays individual data points as markers on a two-dimensional plane. It is useful for examining the relationship between two variables and identifying patterns or clusters in the data.

 *  **Bar Plot:**
> A bar plot represents categorical data using rectangular bars. It is commonly used to compare and visualize the distribution or frequency of different categories or groups.

 * **Histogram:**
> A histogram is used to represent the distribution of a continuous variable. It divides the variable into bins and displays the frequency or proportion of data points falling into each bin.

 * **Pie Chart:**
>  A pie chart is a circular chart divided into sectors, where each sector represents a category or proportion of a whole. It is useful for displaying the relative sizes or percentages of different categories in a dataset.








### **Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**

```
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

### **Note: Also add title, xlabel, and ylabel to the plot.**

### ***ANSWER :***


* **Scatter Plot :**
> A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane. It is particularly useful for examining the relationship between two variables and identifying patterns or clusters in the data.

````python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

#Plot with label
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title("Scatter plot")
plt.scatter(x , y)
````

* **Significance of subplot :**

> The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to create multiple plots arranged in rows and columns, making it easier to compare and visualize different sets of data.

````python
import numpy as np
import matplotlib.pyplot as plt

````

````python
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
````

````python
#Create Sub plots
plt.subplot(2,2,1)
plt.title("Line 1")
plt.xlabel('x')
plt.ylabel('y1')
plt.plot(x,y1 , c = 'b')

plt.subplot(2,2,2)
plt.title("Line 2")
plt.xlabel('x')
plt.ylabel('y2')
plt.plot(x,y2 , c = 'r')

plt.subplot(2,2,3)
plt.title('Line 3')
plt.xlabel('x')
plt.ylabel('y3')
plt.plot(x,y3 , c = 'y')

plt.subplot(2,2,4)
plt.title('Line 4')
plt.xlabel('x')
plt.ylabel('y4')
plt.plot(x,y4 , c = 'g')

plt.tight_layout()
plt.show()

````

### **Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**
```
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```
### ***ANSWER :***

* **Bar Plot :**
> A bar plot is a type of plot that uses rectangular bars to represent categorical data. It is commonly used to compare and visualize the distribution or frequency of different categories or groups.
  
  > Bar plots are used to present categorical data in a clear and visually appealing manner. They provide a way to display and compare the values of different categories or groups, making it easy to identify patterns, trends, and relationships within the data. Bar plots are particularly effective in situations where the data can be divided into distinct categories and the focus is on comparing the values associated with those categories.

* **Here are some key reasons why bar plots are used:**

1. Comparison: Bar plots allow for easy comparison of values between different categories or groups. The lengths of the bars represent the values being compared, making it visually intuitive to understand and interpret the data.

2. Visualizing distributions: Bar plots can show the distribution or frequency of data within each category. The height of each bar represents the value or frequency associated with the category, providing a visual representation of the distribution.

3. Highlighting patterns and trends: Bar plots can reveal patterns, trends, or disparities between categories or groups. They make it easier to identify variations or differences in values across different categories and can provide insights into relationships or comparisons.

4. Communication and presentation: Bar plots are commonly used in reports, presentations, and data visualization to effectively communicate information. They offer a visually appealing and concise way to present categorical data to a wide audience.


````python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
colors = ['blue','orange','yellow','red']

#Plot  verical bar plots with colors
plt.subplot(2,1,1)
plt.bar(company,profit , color = colors)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Vertical bar plot (Company Vs profit)')


#Plot  horizental bar plots with colors
plt.subplot(2,1,2)
plt.barh(company,profit , color = colors)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Horizental bar plot (Company Vs profit)')

plt.tight_layout()
plt.show()
````

### **Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**
```
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

### ***ANSWER :***

* **Box plot :**

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. It provides a summary of the data's central tendency, spread, and any potential outliers.
   * A box plot consists of a box and whiskers. The box represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median. The whiskers extend from the box and show the range of the data, excluding any outliers. Outliers, if present, are plotted individually as points outside the whiskers.

* **Significance of box plot :**

Box plots are useful for comparing distributions and identifying differences in medians, spreads, and potential outliers between different groups or datasets. They provide a concise visual summary of the data's distribution and can reveal insights into the variability and skewness of the dataset.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the box plot
ax.boxplot([box1, box2])

# Set labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot')

# Show the plot
plt.show()
````
